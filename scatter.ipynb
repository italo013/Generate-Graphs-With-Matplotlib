```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Gerando os Dados dos Vetores

```python
x = np.arange(0, 125, 5) #Gerando dados para o Vetor x
y = np.random.random(len(x)) #Gerando dados para o Vetor y
y2 = np.random.uniform(-1, 0, len(x))
display(x, y, y2)
```

Criando o Gr√°fico

```python
plt.figure(figsize=(10,5)) #Definindo o Tamanho do Gr√°fico (Largura, Altura)
plt.scatter(x=x, #Valores do Eixo x
            y=y, #Valores do Eixo y
            color='orange', #Cor dos Marcadores
            marker='h', #Formato dos Marcadores
            s=150, #Tamanho dos Marcadores
            edgecolors='black', #Cor da Borda dos Marcadores
            label='S√©rie 1', #Definindo o nome da Legenda
            linewidths=2) #Espessura dos Marcadores

#Gerando o 2¬∫ Vetor de y
plt.scatter(x=x, #Valores do Eixo x
            y=y2, #Valores do Eixo y
            color='blue', #Cor dos Marcadores
            marker='s', #Formato dos Marcadores
            s=150, #Tamanho dos Marcadores
            edgecolors='black', #Cor da Borda dos Marcadores
            label='S√©rie 2', #Definindo o nome da Legenda
            linewidths=2) #Espessura dos Marcadores

plt.xlabel('x', fontsize=16) #Nomeando o Eixo x
plt.ylabel('y', fontsize=16) #Nomeando o Eixo y
plt.title('Gr√°fico de Dispers√£o', fontsize=25) #Nomeando o T√≠tulo do Gr√°fico
plt.xlim([-1, 125]) #Definindo os Limites do Eixo x
plt.ylim([-1, 1.5]) #Definindo os Limites do Eixo y
plt.legend(ncol=2, loc=0, shadow=True) #Gerando o quadro de Legendas
plt.grid(True) #Criando as grade
plt.savefig('Gr√°fico de Dispers√£o.png', dpi=600, transparent=True) #Salvando o Gr√°fico em png, com a Resolu√ß√£o de 600 dpi e Fundo Transparente
plt.show() #Plotando o Gr√°fico
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mtick

data = [8,12,15,17,18,18.5]
perc = np.linspace(0,100,len(data))

fig = plt.figure(1, (7,4))
ax = fig.add_subplot(1,1,1)

ax.plot(perc, data)

fmt = '%.0f%%' # Format you want the ticks, e.g. '40%'
yticks = mtick.FormatStrFormatter(fmt)
ax.yaxis.set_major_formatter(yticks)

plt.show()
```

```python
x = [1, 2, 3, 4] 
y = [0.1, 0.2, 0.3, 0.4] 
      
fig, (axes1, axes2, axis3) = plt.subplots(1, 3, figsize=(15,5)) 
axes1.plot(x, y)  
axes1.set_ylabel('y-axis')  
axes1.set_xlabel('x-axis') 
axes1.set_xticks(np.arange(0, 5, 1)) 
axes1.set_yticks(np.arange(0, 0.5, 0.1)) 
axes2.plot(x, y)  
axes2.set_ylabel('y-axis')  
axes2.set_xlabel('x-axis') 
axes2.set_xticks(np.arange(0, 4.5, 0.5)) 
axes2.set_yticks(np.arange(0, 0.45, 0.05)) 
  
  
fig.suptitle('set_xticks() and set_yticks() Example') 
fig.tight_layout(pad=3.0) 
  
plt.show()
```
